```python
import calendar

def get_time_from_string_oct(row):
    try:
        hour, minute = row.split(":")
        return int(hour) + int(minute) / 60
    except AttributeError:
        return "checklater"
    
def get_dow_from_date_oct(row):
    try:
        weekday_int =  datetime.datetime.strptime(row, "%m/%d/%Y").weekday()
        return calendar.day_name[weekday_int]
    except:
        return "checklater"
    
# clean_march = mar_19_df.query("numeric_time_val != 'checklater'").copy()
clean_oct = oct_18_df.copy()
clean_oct['numeric_time_val_oct'] = clean_oct['Time Clear'].map(get_time_from_string_oct)
clean_oct['dow'] = clean_oct['Date Clear'].map(get_dow_from_date_oct)

clean_oct.head()
```

```python
# useful converter function gonig from int to or readable time string

def get_back_to_string_time_oct(number):
    if number < 13:
        part1 = (number) * 60
    else:
        part1 = (number - 12) * 60
    return part1

# get_back_to_string_time(14.633333)
oct_avg = clean_oct[clean_oct["numeric_time_val_oct"] != "checklater"]
# clean_march[clean_march["numeric_time_val"]!="checklater"].groupby(['HOSPITAL', 'dow'])["numeric_time_val"].mean().unstack('dow', fill_value=0)
# clean_march[clean_march["numeric_time_val"]!="checklater"].groupby(['HOSPITAL', 'dow']).agg({"numeric_time_val":"mean"})
oct_avg["numeric_time_val_oct"] = oct_avg["numeric_time_val_oct"].apply(lambda x: float(x))
oct_avg.groupby(["Hospital","dow"]).agg({"numeric_time_val_oct":"mean"}).unstack('dow', fill_value=0)

```
