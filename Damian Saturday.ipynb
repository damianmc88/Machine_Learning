```python
def get_time_from_string_jan(row):
    try:
        hour, minute = row.split(":")
        return int(hour) + int(minute) / 60
    except AttributeError:
        return "checklater"
    
def get_dow_from_date_jan(row):
    try:
        weekday_int =  datetime.datetime.strptime(row, "%m/%d/%Y").weekday()
        return calendar.day_name[weekday_int]
    except:
        return "checklater"

clean_jan = jan_19_df.copy()
clean_jan['numeric_time_val_jan'] = clean_jan['Time Clear'].map(get_time_from_string_jan)
clean_jan['dow'] = clean_jan['Date Clear'].map(get_dow_from_date_jan)

clean_jan.head()
```

```python
# useful converter function gonig from int to or readable time string
def get_back_to_string_time_jan(number):
    if number < 13:
        part1 = number
    else:
        part1 = number - 12
    return part1

clean_jan = clean_jan.groupby(['Hospital', 'dow']).mean().unstack('dow', fill_value=0).reset_index()
clean_jan.head()
```

```python
# clean_jan = clean_jan.groupby(['Hospital', 'dow']).mean().unstack('dow', fill_value=0).reset_index()
clean_jan.columns = ("Hospital","Friday","Monday","Saturday","Sunday","Thursday","Tuesday","Wednesday")
clean_jan
```

```python
def get_back_to_real_time_jan(number):
    if number >= 12:
        ap = "PM"
    else:
        ap = "AM"
    hour = (int(number)-1) % 12 + 1
    minute = str(round((number - int(number)) * 60))
    if len(minute) == 1:
        minute = "0" + str(minute)
    return f"{hour}:{minute} {ap}"

for day in ['Friday', 'Monday', 'Saturday', 'Sunday', 'Thursday','Tuesday', 'Wednesday']:
    clean_jan[day] = clean_jan[day].apply(lambda x: get_back_to_real_time_jan(x))
clean_jan
```
